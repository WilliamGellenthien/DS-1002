```python
import numpy as np
```

```python
from skimage import io

photo = io.imread('/content/Stock Image for PrgAssignment.jpeg')
```

```python
def red_filter(photo):
    red_filtered_photo = np.copy(photo)

    red_filtered_photo[:,:,1] = 0  # Setting green channel to zero
    red_filtered_photo[:,:,2] = 0  # Setting blue channel to zero

    return red_filtered_photo
```

```python
red_result = red_filter(photo)
```

```python
import matplotlib.pyplot as plt     # I did this to be able to show the pictures
```

```python
def green_filter(photo):
    green_filtered_photo = np.copy(photo)

    # Set red and blue channels to zero
    green_filtered_photo[:,:,0] = 0
    green_filtered_photo[:,:,2] = 0

    return green_filtered_photo
```

```python
def blue_filter(photo):
    blue_filtered_photo = np.copy(photo)

    # Set red and green channels to zero
    blue_filtered_photo[:,:,0] = 0
    blue_filtered_photo[:,:,1] = 0

    return blue_filtered_photo
```

```python
from PIL import Image
url = input("/content/Stock Image for PrgAssignment.jpeg")
#When it asks you for the input you go to where the file would be uploaded and then hit copy path
# and then paste that into the input box

# This is for loading the image
photo = Image.open(url)

# This is for converting the image to an array
original_photo = np.array(photo)

# This is applying the filters
red_result = red_filter(original_photo)
green_result = green_filter(original_photo)
blue_result = blue_filter(original_photo)

# This is combining channels
def combine_channels(red, green, blue):
    combined_photo = np.stack([red[:, :, 0], green[:, :, 1], blue[:, :, 2]], axis=-1)
    return combined_photo
combined_rg = combine_channels(red_result, green_result, green_result)
combined_rb = combine_channels(red_result, blue_result, blue_result)
combined_gb = combine_channels(green_result, green_result, blue_result)
        # For the combined channels above, it did not matter which one of the "results" I put into the slot of
        #color I did not want.  See how I put two green_result in the first combine channel, it did not matter if
        #I put red_result or green_result in the blue_result space, the picture would look the same


# This is to display the images
plt.figure(figsize=(15, 8))

plt.subplot(2, 4, 1)
plt.imshow(original_photo)
plt.title('Original')

plt.subplot(2, 4, 2)
plt.imshow(red_result)
plt.title('Red Filtered')

plt.subplot(2, 4, 3)
plt.imshow(green_result)
plt.title('Green Filtered')

plt.subplot(2, 4, 4)
plt.imshow(blue_result)
plt.title('Blue Filtered')

plt.subplot(2, 4, 5)
plt.imshow(combined_rg)
plt.title('R + G, No B')

plt.subplot(2, 4, 6)
plt.imshow(combined_rb)
plt.title('R + B, No G')

plt.subplot(2, 4, 7)
plt.imshow(combined_gb)
plt.title('G + B, No R')

```

```python
# This is the check
is_identical = np.array_equal(original_photo, combine_channels(red_result, green_result, blue_result))

plt.subplot(2, 4, 8)
plt.text(0.5, 0.5, f'Identical: {is_identical}', fontsize=12, ha='center')
plt.axis('off')

plt.show()
```
